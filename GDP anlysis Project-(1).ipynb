ANALYSING ARAB WORLD

```python
df_pr=df[df["Country Name"]=="Arab World"]
```

```python
df_pr
```

```python
#1968-2576
#1969-2843
#1970-3138
```

```python
round(((3138-2843)/2843)*100,2)
```

```python
data=df_pr.values
gdp_change=[0]
for i in range(1,len(data)):
     prev=data[i-1][3]
     cur=data[i][3]
     gdp_change.append(round(((cur-prev)/prev)*100,2))
```

```python
df_pr.assign(GDP=gdp_change)
```

```python

df_pr.plot(kind="line",x="Year",y="Value",grid=True,legend=False,ylabel="GDP",xlabel="YEARS")
```
